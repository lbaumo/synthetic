```python
scat = fio.read('test2_prepsf.cat', ext=2)
```

```python
number = scat['NUMBER']
alpha_j2000 = scat["ALPHA_J2000"]
delta_j2000 = scat["DELTA_J2000"]
a_image = scat["A_IMAGE"]
```

```python
obj_data = meds.util.get_meds_input_struct(len(scat)) 
obj_data['id'] = number
obj_data['ra'] = alpha_j2000
obj_data['dec'] = delta_j2000
obj_data['box_size'] = np.round(a_image*20)
```

```python
# image_file='MACS1447_V190110_r_8.fits'
image_file ='i.MACS1447.fits'
lenght_image_file = len(image_file)
image_info = meds.util.get_image_info_struct(1,lenght_image_file)
image_info['image_path'][0] = image_file
image_info['magzp'][0] = 30
```

```python
maker=meds.MEDSMaker(obj_data,image_info)
```

```python
maker.write('testmeds.fits')
```
